# Python Data Structures
In this section, we will delve into Python's basic data structures and understand their SAS equivalents. Understanding these data structures is fundamental for data wrangling and data linkage tasks.

## Lists

**Python**

A list in Python is an ordered collection of items which can be of any type. Lists are mutable, meaning the values in a list can be changed.

```python
# Creating a list in Python
my_list = [1, 'apple', 3.5]

# Accessing elements
first_element = my_list[0]  # Output: 1
```
**SAS**
In SAS, the closest equivalent to a Python list would be an array. However, SAS arrays are not as flexible as Python lists; they must contain elements of the same type.
```
DATA work.array_example;
  array fruits[3] $ ('apple', 'banana', 'cherry');
RUN;
```


## Dictionaries

**Python**

A dictionary is an unordered collection of key-value pairs.
```python
# Creating a dictionary in Python
my_dict = {'apple': 1, 'banana': 2, 'cherry': 3}

# Accessing values by their key
apple_count = my_dict['apple']  # Output: 1
```

**SAS**

SAS doesn't have a direct equivalent to Python dictionaries. The closest would be using a combination of data steps and PROC steps to simulate key-value pair operations.

## Tuples
**Python**

A tuple is similar to a list but is immutable.

```python
# Creating a tuple in Python
my_tuple = (1, 'apple', 3.5)

# Accessing elements
first_element = my_tuple[0]  # Output: 1
```

**SAS**

Again, there isn't a direct SAS equivalent to a Python tuple.



## Sets
**Python**

A set is an unordered collection of unique items.
```python
# Creating a set in Python
my_set = {1, 'apple', 3.5, 'apple'}

# The duplicate 'apple' is removed
# Output: {1, 3.5, 'apple'}
```

**SAS**

SAS does not have a direct equivalent to Python sets


## Key Differences Between SAS and Python Data Structures
* **Flexibility**: Python data structures like lists and dictionaries are more flexible than SAS arrays or data step variables.
* **Type Handling:** Python data structures can store multiple types of elements, unlike SAS arrays.
* **Syntax:** Python data structures often require fewer lines of code and less boilerplate than SAS equivalents.
* **Immutable vs Mutable:** Python provides both mutable (list, dictionary, set) and immutable (tuple) options, whereas SAS data structures are generally mutable.
* **Built-in Operations:** Python has a wealth of built-in methods for manipulating these data structures, providing a more streamlined workflow for data wrangling tasks.

By understanding these differences, you'll be well-equipped to write Python code effectively, especially for data manipulation tasks.

For more on lists and other data structures, check out the [official documentation ](https://docs.python.org/3/tutorial/datastructures.html)

# Data Wrangling with Pandas

In this section, we will focus on how to manipulate and transform data using Pandas, similar to some of the steps you are familiar with using SAS. Understanding this will be essential for tasks like data cleaning, preparation, and transformation for analysis.

First, let's import the Pandas library and create some example data to work with.

````python
# Importing Pandas library
import pandas as pd

# Creating example DataFrame
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40],
    "Department": ["HR", "Finance", "Engineering", "Marketing"]
})

df
````

## Data Filtering

Let's filter out the rows where the age is greater than 30.

````python
# Filtering based on condition
filtered_df = df[df['Age'] > 30]
filtered_df
````

## Column Operations

We can create a new column based on existing columns. Let's create a column that indicates the age of individuals next year.


````python
df['Age_Next_Year'] = df['Age'] + 1
df
````

### SAS Equivalent

```
DATA work.new_data;
  SET work.sample_data;
  Age_Next_Year = Age + 1;
RUN;
```

## Aggregation - Group by and mean operation
We can perform aggregation to summarize the data. Let's find the average age by department.


````python
grouped_df = df.groupby("Department")["Age"].mean()
grouped_df
````

### SAS Equivalent

```
PROC SQL;
  CREATE TABLE work.grouped_data AS
  SELECT Department, SUM(Salary) AS Total_Salary
  FROM work.sample_data
  GROUP BY Department;
QUIT;
```

## Joining Tables

Joining tables combines rows from two or more tables based on a related column. Let's create another DataFrame and perform an inner join.

````python
# Creating another example DataFrame
df2 = pd.DataFrame({
    "Name": ["Alice", "Bob", "Eve"],
    "Salary": [50000, 60000, 70000]
})

# Performing an inner join on 'Name'
merged_df = pd.merge(df, df2, on='Name')
merged_df
````

### SAS Equivalent
```
PROC SQL;
  CREATE TABLE work.merged_data AS
  SELECT *
  FROM work.table1 A, work.table2 B
  WHERE A.Key_Column = B.Key_Column;
QUIT;
```


## Sorting

Sorting arranges the data in a particular order. Let's sort the DataFrame by the 'Age' column.


````python
# Sorting DataFrame by 'Age'
sorted_df = df.sort_values(by='Age')
sorted_df
````

### SAS Equivalent
```
PROC SORT DATA=work.sample_data;
  BY Age;
RUN;
```

## Missing Data

Handling missing data is an essential step in the data wrangling process. Let's introduce some missing values and then fill them.


````python
# Introducing missing values
df.loc[1, 'Age'] = None
df.loc[3, 'Age'] = None

# Filling missing values
df.fillna(value={"Age": df['Age'].mean()}, inplace=True)
df
````

### SAS Equivalent
```
DATA work.cleaned_data;
  SET work.sample_data;
  IF MISSING(Age) THEN Age = 0;
RUN;
```

## Pandas DataFrame: Filter by Timestamp using TimeDelta string

````python
df = pd.DataFrame({
  'time': ['2022-09-14 00:52:00-07:00', '2022-09-14 00:52:30-07:00',
           '2022-09-14 01:52:30-07:00'],
  'letter': ['A', 'B', 'C'],
})
df['time'] = pd.to_datetime(df.time)

def rows_in_time_range(df, time_column, start_ts_str, timedelta_str):
  # Return rows from df, where start_ts < time_column <= start_ts + delta.
  # start_ts_str can be a date '2022-09-01' or a time '2022-09-14 00:52:00-07:00'
  # timedelta_str examples: '2 minutes'  '2 days 2 hours 15 minutes 30 seconds'
  start_ts = pd.Timestamp(start_ts_str).tz_localize('US/Pacific')
  end_ts = start_ts + pd.to_timedelta(timedelta_str)
  return df.query("@start_ts <= {0} < @end_ts".format(time_column))

rows_in_time_range(df, 'time', '2022-09-14 00:00', '52 minutes 31 seconds')
````

## Pandas DataFrame: Reshape to have 1 row per value in a list column

Creates a new DataFrame that is a transformed version of the input. E.g.
*   Input: df with a column named `msg_ids` that is a list of values (i.e. many per row, at least in some rows).
*   Output: new_df which has 1 row per unique value found in any of the original `msg_ids` lists, with that value in a new column named `msg_id`.


````python
df = pd.DataFrame({
  'date': ['9/1/22', '9/2/22', '9/3/22'],
  'action': ['Add', 'Update', 'Delete'],
  'msg_ids': [[1, 2, 3], [], [2, 3]],
})
df.set_index('date', inplace=True)


temp_series = df['msg_ids'].apply(pd.Series, 1).stack()
temp_series.index = temp_series.index.droplevel(-1)
temp_series.name = 'msg_id'
new_df = temp_series.to_frame()
new_df.set_index('msg_id', inplace=True)
new_df.loc[~new_df.index.duplicated(), :] # Drop duplicates.
````

## Pandas: DataFrames: Group Timeseries by Frequency

You can group timestamped data into intervals of arbitrary duration using a Grouper object to specify groupby instructions.  The `freq` parameter is a string that may contain an integer followed by an [offset alias](https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#offset-aliases).  E.g. To see output for 2 minute long intervals:

````python
df = pd.DataFrame({
  'time': ['2022-09-01 00:00:01-07:00', '2022-09-01 00:00:02-07:00',
           '2022-09-01 00:01:00-07:00', '2022-09-01 00:02:00-07:00',
           '2022-09-01 00:03:00-07:00', '2022-09-01 00:04:00-07:00',
           '2022-09-01 00:05:00-07:00', '2022-09-01 00:07:00-07:00'],
  'requests': [1, 1, 1, 1, 1, 1, 1, 1],
})
df['time'] = pd.to_datetime(df.time)

df.groupby(pd.Grouper(key='time', freq='2min')).sum()
````
